```python
import numpy as np
import pandas as pd

# Calculate descriptive statistics for numeric columns
df = dataset_tracks.select_dtypes(include=[np.number])
stats = {
    
    'range': df.max() - df.min(),
    'variance': df.var(),
    'coeff_var': df.std() / df.mean(),
    'q1': df.quantile(0.25),
    'q3': df.quantile(0.75),
    'iqr': df.quantile(0.75) - df.quantile(0.25),
    'skew': df.skew(),
    'kurtosis': df.kurtosis(),
    'mode': df.mode().iloc[0],
    'median': df.median(),
    'mid_range': (df.max() + df.min()) / 2,
    'q_dispersion': (df.quantile(0.75) - df.quantile(0.25)) / (df.quantile(0.75) + df.quantile(0.25)),
    'geometric_mean': np.exp(np.log(df[df > 0]).mean())
}

# Convert dictionary to pandas dataframe and display as table
stats_df = pd.DataFrame(stats)
stats_df.index.name = 'Variable'
stats_df.reset_index(inplace=True)
stats_df.style.hide_index()
```

<div class="alert alert-block alert-success">
<b><h3>Chebyshev's theorem </h3> is a statistical rule that states that for any given dataset, the proportion or percentage of the observations that fall within k standard deviations of the mean is always at least 1 - 1/k^2, where k is any positive integer greater than 1. In other words, it provides a lower bound on the proportion of the data that falls within a certain distance from the mean.</b>
</div>

```python
# Calculate the mean and standard deviation of the danceability column
danceability_mean = dataset_tracks['danceability'].mean()
danceability_std = dataset_tracks['danceability'].std()

# Calculate the minimum and maximum values based on Chebyshev's Theorem
k = 3  # number of standard deviations
min_danceability = danceability_mean - k * danceability_std
max_danceability = danceability_mean + k * danceability_std

# Calculate the percentage of data within k standard deviations of the mean
within_k_std_dev = (dataset_tracks['danceability'] >= min_danceability) & (dataset_tracks['danceability'] <= max_danceability)
percent_within_k_std_dev = 100 * len(dataset_tracks[within_k_std_dev]) / len(dataset_tracks)

print(f"The percentage of data within {k} standard deviations of the mean is {percent_within_k_std_dev:.2f}%")
```

<div class="alert alert-block alert-success">
    <i><b>The percentage of data within 3 standard deviations of the mean is 100.00%</b></i>
</div>

<div class="alert alert-block alert-success">
    <b>Applying the Empirical rule</b>
</div>

```python
# Calculate mean and standard deviation of duration_ms
mean_duration = np.mean(dataset_tracks['duration_ms'])
std_duration = np.std(dataset_tracks['duration_ms'])

# Calculate the range of values that fall within 1, 2, and 3 standard deviations from the mean
range_1std = (mean_duration - std_duration, mean_duration + std_duration)
range_2std = (mean_duration - 2*std_duration, mean_duration + 2*std_duration)
range_3std = (mean_duration - 3*std_duration, mean_duration + 3*std_duration)

# Calculate the proportion of data that falls within each range
prop_within_1std = len(dataset_tracks[(dataset_tracks['duration_ms'] >= range_1std[0]) & (dataset_tracks['duration_ms'] <= range_1std[1])]) / len(dataset_tracks)
prop_within_2std = len(dataset_tracks[(dataset_tracks['duration_ms'] >= range_2std[0]) & (dataset_tracks['duration_ms'] <= range_2std[1])]) / len(dataset_tracks)
prop_within_3std = len(dataset_tracks[(dataset_tracks['duration_ms'] >= range_3std[0]) & (dataset_tracks['duration_ms'] <= range_3std[1])]) / len(dataset_tracks)

# Print the results
print("Approximately {:.2%} of the data falls within 1 standard deviation from the mean.".format(prop_within_1std))
print("Approximately {:.2%} of the data falls within 2 standard deviations from the mean.".format(prop_within_2std))
print("Approximately {:.2%} of the data falls within 3 standard deviations from the mean.".format(prop_within_3std))
```
